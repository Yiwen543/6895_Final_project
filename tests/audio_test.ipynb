# Audio Tests

Requires a microphone. Run `main.ipynb` first, or bootstrap in Cell 1.


```python
# Cell 1: Bootstrap
import sys, os

PROJECT_ROOT = os.path.abspath(os.path.join(os.path.dirname("audio_test.ipynb"), ".."))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)
print("Project root:", PROJECT_ROOT)

from sentence_transformers import SentenceTransformer
from llm_parser import LLMParser
from memory import MemoryManager
from agent import NovaAgent
from audio import STTModel, TTSEngine, AudioListener

_embed_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
embed = lambda t: _embed_model.encode(t, convert_to_numpy=True).tolist()

llm      = LLMParser()
stt      = STTModel()
tts      = TTSEngine()
memory   = MemoryManager(embed_fn=embed)
nova     = NovaAgent(llm=llm, memory=memory, speak=tts.speak)
listener = AudioListener(agent=nova, stt=stt)
print("Audio test ready.")
```

```python
# Cell 2: Single-round test (3-second clip)
# Say something after the prompt appears.
result = listener.run_one_round(duration_sec=3)
print("\nResult:", result)
```

```python
# Cell 3: Continuous VAD loop
# Speak naturally; Nova captures full utterances automatically.
# Press Ctrl+C to stop.
listener.continuous_loop()
```
